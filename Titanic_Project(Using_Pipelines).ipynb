```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv("/content/drive/MyDrive/train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Ticket','Name','Cabin'],inplace =True)
```

```python
df.head()
```

```python
X_train , X_test , y_train , y_test = train_test_split(df.drop(columns=['Survived']),df['Survived'], test_size =0.2, random_state = 42)
```

```python
df.isnull().sum()
```

```python
#Filling missing data using Simple Imputer
#missing data : age (highest no of missing values) : fill missing values with mean of age
#Embarked : fill missing data with most-frequent value
si_age = SimpleImputer()
si_embarked = SimpleImputer(strategy="most_frequent")

X_train_age = si_age.fit_transform(X_train[['Age']])
X_test_age = si_age.transform(X_test[['Age']])

X_train_embarked = si_embarked.fit_transform(X_train[['Embarked']])
X_test_embarked = si_embarked.transform(X_test[['Embarked']])
```

```python
import sklearn
print(sklearn.__version__)
from sklearn.preprocessing import OneHotEncoder
print(OneHotEncoder.__init__.__doc__)
```

```python
#Using One Hot Encoding on nominal data : Embarked and Sex
ohe_embarked = OneHotEncoder(sparse_output=False ,handle_unknown='ignore')
ohe_sex = OneHotEncoder(sparse_output=False,handle_unknown='ignore')

X_train_embarked = ohe_embarked.fit_transform(X_train_embarked)
X_test_embarked = ohe_embarked.transform(X_test_embarked)

X_train_sex = ohe_sex.fit_transform(X_train[['Sex']])
X_test_sex = ohe_sex.transform(X_test[['Sex']])
```

```python
X_train.head()
```

```python
X_train_rem = X_train.drop(columns=['Sex','Age','Embarked'])
X_test_rem = X_test.drop(columns=['Sex','Age','Embarked'])
```

```python
X_train_transformed = np.concatenate((X_train_rem,X_train_age,X_train_embarked,X_train_sex),axis=1)
X_test_transformed = np.concatenate((X_test_rem,X_test_age,X_test_embarked,X_test_sex),axis=1)
```

```python
X_test_transformed.shape
```

```python
dr = DecisionTreeClassifier()
dr.fit(X_train_transformed,y_train)
```

```python
y_pred = dr.predict(X_test_transformed)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(ohe_sex,open('/content/drive/MyDrive/models/ohe_sex.pkl','wb'))
pickle.dump(ohe_embarked,open('/content/drive/MyDrive/models/ohe_embarked.pkl','wb'))
pickle.dump(dr,open('/content/drive/MyDrive/models/dr.pkl','wb'))
```
